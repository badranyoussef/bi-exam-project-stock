```python
from polygon import RESTClient
import pandas as pd
```

```python
# Initializing RESTClient
client = RESTClient("zgHnQp9h1GsoDYXtmwmvQ8NAWX4GkJ9T")

# Retrieveing data from API
aggs = []
for a in client.list_aggs(
    "AAPL",
    1,
    "minute",
    "2022-01-01",
    "2023-02-03",
    limit=50000,
):
    aggs.append(a)
 
apple_df = pd.DataFrame(aggs)
```
