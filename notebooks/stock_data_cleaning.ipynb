```python
# removing the dollar symbol from price at different columns and konvert til float

# columns with dollar symbol
price_columns = ['Close/Last', 'Open', 'High', 'Low']

# function til remove the symbol
def remove_dollarsymbol(df, price_columns):
    for column in price_columns:
        df[column] = df[column].str.replace('$', '', regex=False).astype(float)
    return df
```

```python
appl_df = remove_dollarsymbol(appl_df, price_columns)
amzn_df = remove_dollarsymbol(amzn_df, price_columns)
msft_df = remove_dollarsymbol(msft_df, price_columns)
```
